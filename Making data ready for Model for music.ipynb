```python
# now making inout ready for model
network_input=[]
network_output=[]
# as we are currently reading in string form
for i in range(len(notes)-sequence_length):
    seq_inp=notes[i:i+sequence_length]
    seq_out=notes[i+sequence_length]
    
    # now converting string to int by dictionary
    
    network_input.append([ele_to_int[ch] for ch in seq_inp])
    network_output.append(ele_to_int[seq_out])
    
    
```

```python
# now making desired shape of input data to fit in model  3-d  like   (sample_size,sequence_length,__)
# as now network_input is 2D list 
import numpy as np
network_input=np.reshape(network_input,(n_paterns,sequence_length,1))
```

```python
# now normalising the data
from keras.utils import np_utils
```

```python
# preparing output data for trainig
network_output=np_utils.to_categorical(network_output)
```

```python
# normalising input data   
# as we have 359 unique integer(after encoding) so converting into one hot 

normalised_network_input=network_input/float(n_vocabs)
```

# Creating Model

```python
from keras.layers import *
from keras.models import Sequential,load_model
from keras.callbacks import ModelCheckpoint, EarlyStopping
```

```python
model=Sequential()
model.add(LSTM(512,input_shape=(normalised_network_input.shape[1],normalised_network_input.shape[2]),
               return_sequences=True))
model.add(Dropout(0.3))
model.add(LSTM(512,return_sequences=True))
model.add(Dropout(0.3))
model.add(LSTM(512))
model.add(Dense(256))
model.add(Dropout(0.3))
model.add(Dense(n_vocabs,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss="categorical_crossentropy", optimizer="adam")
```

```python
checkpoint=ModelCheckpoint("model.hdf5", monitor='loss', verbose=0, save_best_only=True, mode='min')
model_his=model.fit(normalised_network_input,network_output,epochs=2,batch_size=64,callbacks=[checkpoint])
```
